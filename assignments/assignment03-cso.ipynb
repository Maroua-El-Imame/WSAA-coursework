This program retrieves the dataset for the "exchequer account (historical series)" from the CSO.  
It then stores the data into a file called "cso.json" in the WSAA assignments repository. 

```python
import requests
import json
```

```python
# API url

url = "https://ws.cso.ie/public/api.restful/PxStat.Data.Cube_API.ReadDataset/FIQ02/JSON-stat/2.0/en"
```

```python
# function that sends requests to CSO api to retrieve data and convert the response into JSON format

def get_exchequer_data():
    response = requests.get(url)
    return response.json()
```

```python
# save data locally as json file in a readable (indented) formt

if __name__ == "__main__":
    with open("cso.json", "w") as f:
        json.dump(get_exchequer_data(), f, indent=4)
```
